```python
import pandas as pd
df = pd.read_excel('Tarifas por zonas 2016-2017.xlsx')
```

```python
# tabla completa
df.head(72)
```

```python
# tabla 2016

tabla2016 = df.head (40)
tabla2016
```

```python
# tabla 2017

tabla2017 = df.tail(32)
tabla2017
```

```python
# promedio ventas 2016

promedioVentas2016 = tabla2016.uso_base_firme
print(promedioVentas2016.mean())
```

```python
#promedio ventas 2017

promedioVentas2017 = tabla2017.uso_base_firme
print(promedioVentas2017.mean())
```

```python
# variaciÃ³n ventas 2016

variacionVentas2016 = tabla2016.uso_base_firme
print(variacionVentas2016.var())
```

```python
# variaciÃ³n ventas 2017

variacionVentas2017 = tabla2017.uso_base_firme
print(variacionVentas2017.var())
```

```python
# demanda promedio ventas 2016

demandaPromedio2016 = tabla2016.volumetrica
print(demandaPromedio2016.mean())
```

```python
# demanda promedio ventas 2017

demandaPromedio2017 = tabla2017.volumetrica
print(demandaPromedio2017.mean())
```

```python
# valor mÃ¡ximo tarifa 2016
valorMaximo2016 = tabla2016.uso_base_firme
print(valorMaximo2016.max())
```

```python
# valor mÃ¡ximo tarifa 2017
valorMaximo2017 = tabla2017.uso_base_firme
print(valorMaximo2017.max())
```

# Genera un grÃ¡fico de lÃ­nea para visualizar la variaciÃ³n de las tarifas contra el tiempo para los dos aÃ±os.

```python
#variacion de tarifas 2016

df_2016 = pd.DataFrame(tabla2016)
fecha2016 = df_2016.fecha_inicio

df_2016['fecha2016'] = pd.to_datetime(df_2016['fecha_inicio'])
df_2016.set_index('fecha2016', inplace=False)

plt.bar(df_2016['fecha_inicio'], df_2016['minima_base_interrumpible'])

plt.show()
```

```python
#variacion de tarifas 2017

df_2017 = pd.DataFrame(tabla2017)
fecha2017 = df_2017.fecha_inicio

df_2017['fecha2017'] = pd.to_datetime(df_2017['fecha_inicio'])
df_2017.set_index('fecha2017', inplace=False)

plt.bar(df_2017['fecha_inicio'], df_2017['minima_base_interrumpible'])

plt.show()
```

# Genera un grÃ¡fico de barras que seÃ±ale mÃ¡ximos y mÃ­nimos de las tarifas por mes de cada uno de los dos aÃ±os.

```python
import numpy as np
```

```python
#valores de 2016
valorMaximoEnero2016 = max(df_2016.volumetrica[1:9])
valorMinimoEnero2016 = min(df_2016.volumetrica[1:9])

valorMaximoAbril2016 = max(df_2016.volumetrica[10:17])
valorMinimoAbril2016 = min(df_2016.volumetrica[10:17])

valorMaximoMayo2016 = max(df_2016.volumetrica[18:25])
valorMinimoMayo2016 = min(df_2016.volumetrica[18:25])

valorMaximoJulio2016 = max(df_2016.volumetrica[26:33])
valorMinimoJulio2016 = min(df_2016.volumetrica[26:33])

valorMaximoOctubre2016 = max(df_2016.volumetrica[34:41])
valorMinimoOctubre2016 = min(df_2016.volumetrica[34:41])


valoresMaximos = [valorMaximoEnero2016, valorMaximoAbril2016, valorMaximoMayo2016, valorMaximoJulio2016, valorMaximoOctubre2016]
valoresMinimos = [valorMinimoEnero2016, valorMinimoAbril2016, valorMinimoMayo2016, valorMinimoJulio2016, valorMinimoOctubre2016]

valorMaximoEnero2016 = float(valorMaximoEnero2016)
```

```python
#grÃ¡fico de 2016
categorias = ['Enero','Abril', 'Mayo', 'Julio', 'Octubre']

ancho_barra = 0.20  # Ancho de cada barra
indice = np.arange(len(categorias))  # Ãndices para las categorÃ­as

fig, ax = plt.subplots()

ax.bar(indice, valoresMaximos, width=ancho_barra, label='Valores Maximos')
ax.bar(indice + ancho_barra, valoresMinimos, width=ancho_barra, label='Valores Minimos')

ax.set_xlabel('Meses')
ax.set_ylabel('Valores')
ax.set_title('Valores minimos y maximos')
ax.set_xticks(indice + ancho_barra)
ax.set_xticklabels(categorias)
ax.legend()

plt.show()
```

```python
#valores de 2017
valorMaximoEnero2017 = max(df_2017.volumetrica[1:8])
valorMinimoEnero2017 = min(df_2017.volumetrica[1:8])

valorMaximoAbril2017 = max(df_2017.volumetrica[9:16])
valorMinimoAbril2017 = min(df_2017.volumetrica[9:16])


valorMaximoJulio2017 = max(df_2017.volumetrica[17:24])
valorMinimoJulio2017 = min(df_2017.volumetrica[17:24])

valorMaximoOctubre2017 = max(df_2017.volumetrica[25:32])
valorMinimoOctubre2017 = min(df_2017.volumetrica[25:32])


valoresMaximos2017 = [valorMaximoEnero2017, valorMaximoAbril2017, valorMaximoJulio2017,valorMaximoOctubre2017]
valoresMinimos2017 = [valorMinimoEnero2017, valorMinimoAbril2017, valorMinimoJulio2017, valorMinimoOctubre2017]
```

```python
#grÃ¡fico de 2017

categorias2017 = ['Enero','Abril', 'Julio', 'Octubre']

ancho_barra = 0.20  # Ancho de cada barra
indice = np.arange(len(categorias2017))  # Ãndices para las categorÃ­as

fig, ax = plt.subplots()

ax.bar(indice, valoresMaximos2017, width=ancho_barra, label='Valores Maximos')
ax.bar(indice + ancho_barra, valoresMinimos2017, width=ancho_barra, label='Valores Minimos')

ax.set_xlabel('Meses')
ax.set_ylabel('Valores')
ax.set_title('Valores minimos y maximos')
ax.set_xticks(indice + ancho_barra)
ax.set_xticklabels(categorias2017)
ax.legend()

plt.show()
```
